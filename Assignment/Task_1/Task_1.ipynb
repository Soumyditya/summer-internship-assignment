```python
user_input = input("Enter Numbers").split()
numbers = []  # Converting input into list of integers

for i in user_input:
  numbers.append(int (i))


# Removing duplicates and sorting
sorted_numbers = sorted(set(numbers))

# Frequency of elements
frequency = {}
for num in numbers:
  if num in frequency:
    frequency[num] += 1
  else:
    frequency[num] = 1

data = {
    "Maximum Value": max(numbers),
    "Minimum Value": min(numbers),
    "Average Value": sum(numbers) / len(numbers),
    "Frequency": frequency
 }
print("\nSorted List Without Duplicates:")
print(sorted_numbers)

print("\nDictionary Output:")
print(data)
```
