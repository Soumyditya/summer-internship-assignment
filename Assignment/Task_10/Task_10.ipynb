```python
import pandas as pd
import matplotlib.pyplot as plt

# Load collected data
data = pd.read_csv("scraped_data.csv")


# Total Posts Collected


total_posts = len(data)

print("\nTotal Posts Collected:", total_posts)


# Total Unique Creators

if "Author" in data.columns:

    unique_creators = data["Author"].nunique()

    print("Total Unique Creators:", unique_creators)

else:

    unique_creators = 0

# Most Common Keywords


common_keywords = data["Topic"].value_counts()

print("\nMost Common Keywords:\n")

print(common_keywords)


# Top Performing Posts


print("\nTop Posts:\n")

print(data.head())


# Visualization 1 - Keyword Bar Chart


plt.figure(figsize=(8, 5))

plt.bar(common_keywords.index,
        common_keywords.values)

plt.title("Most Common Keywords")

plt.xlabel("Keywords")

plt.ylabel("Count")

plt.savefig("keywords_chart.png")

plt.close()


# Visualization 2 - Total Posts Pie Chart


plt.figure(figsize=(6, 6))

plt.pie(common_keywords.values,
        labels=common_keywords.index,
        autopct="%1.1f%%")

plt.title("Keyword Distribution")

plt.savefig("posts_pie_chart.png")

plt.close()

# Save Summary Report


report = open("summary_report.txt", "w")

report.write("DATA SUMMARY REPORT\n\n")

report.write(f"Total Posts Collected: {total_posts}\n")

report.write(f"Total Unique Creators: {unique_creators}\n\n")

report.write("Most Common Keywords:\n")

report.write(str(common_keywords))

report.close()

print("\nSummary Report Saved Successfully")
```
