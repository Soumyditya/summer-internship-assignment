```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import os

# Read topics from file
file = open("keywords.txt", "r", encoding="utf-8")

topics = [line.strip() for line in file.readlines() if line.strip()]

file.close()

# Create main folder
os.makedirs("videos", exist_ok=True)

# Remove duplicate titles
all_titles = set()

# Store all scraped data
all_data = []


# Function for scraping
def scrape_data(topic):

    print("\nSearching for:", topic)

    # Create folder name
    folder_name = topic.split()[0].lower()

    # Create category folder
    os.makedirs("videos/" + folder_name, exist_ok=True)

    # Search URL
    url = f"https://www.sciencedirect.com/search?qs={topic}"

    # Headers
    headers = {
        "User-Agent": "Mozilla/5.0"
    }

    try:

        # Send request
        response = requests.get(url, headers=headers)

        # Parse HTML
        soup = BeautifulSoup(response.text, "html.parser")

        # Find article titles
        articles = soup.find_all("a")

        count = 0

        for article in articles:

            title = article.text.strip()

            # Skip empty titles
            if title == "":
                continue

            # Remove duplicates
            if title in all_titles:
                continue

            all_titles.add(title)

            # Print title
            print("\nTitle:", title)

            # Save title in text file
            save_file = open(
                "videos/" + folder_name + "/data.txt",
                "a",
                encoding="utf-8"
            )

            save_file.write(title + "\n")

            save_file.close()

            # Store data
            all_data.append({

                "Topic": topic,
                "Title": title

            })

            count += 1

            # Get first 5 results
            if count == 5:
                break

    except Exception as e:

        print("Error:", e)


# Run scraper for all topics
for topic in topics:

    scrape_data(topic)


# Save CSV File


df = pd.DataFrame(all_data)

df.to_csv("scraped_data.csv", index=False)

# Print Final Message


print("\nAll Data Collected Successfully")

print("\nCSV File Created Successfully")
```
