```python
import json
import random

log_entries = [
   "INFO User logged in",
    "ERROR Database timeout",
    "WARNING Memory usage high",
    "ERROR Invalid credentials",
    "INFO File uploaded",
]

# Creating random logs
with open("logs.txt", "w") as file:
    for _ in range(100):
        random_log = random.choice(log_entries)
        file.write(random_log + "\n")

# Read file and count log
log_count = {}
with open("logs.txt", "r") as file:
    for line in file:
        log_type = line.split()[0]

        if log_type in log_count:
            log_count[log_type] += 1
        else:
            log_count[log_type] = 1

# print log count
print("Log Type Count")
for key, value in log_count.items():
    print(f"{key}: {value}")

# Find most frequent log type
max_count = 0
most_frequent = ""

for key, value in log_count.items():
    if value > max_count:
        max_count = value
        most_frequent = key

print("\nMost Frequent Log Type:")
print(f"{most_frequent} occurred {max_count} times")

```
