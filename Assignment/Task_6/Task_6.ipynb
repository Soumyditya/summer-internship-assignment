```python
import pandas as pd
import matplotlib.pyplot as plt

# Load cleaned dataset
data = pd.read_csv("cleaned_dataset.csv")

# -----------------------------------
# 1. Histogram
# -----------------------------------

plt.figure(figsize=(8, 5))

plt.hist(data["productivity_score"], bins=10)

plt.title("Productivity Score Distribution")
plt.xlabel("Productivity Score")
plt.ylabel("Frequency")

plt.savefig("histogram.png")

plt.close()

# -----------------------------------
# 2. Scatter Plot
# -----------------------------------

plt.figure(figsize=(8, 5))

plt.scatter(data["deep_work_hours"],
            data["productivity_score"],
            label="Employees")

plt.title("Deep Work Hours vs Productivity")
plt.xlabel("Deep Work Hours")
plt.ylabel("Productivity Score")

plt.legend()

plt.savefig("scatter_plot.png")

plt.close()

# -----------------------------------
# 3. Heatmap
# -----------------------------------

correlation = data.corr(numeric_only=True)

plt.figure(figsize=(10, 6))

plt.imshow(correlation, cmap="coolwarm")

plt.colorbar()

plt.xticks(range(len(correlation.columns)),
           correlation.columns,
           rotation=90)

plt.yticks(range(len(correlation.columns)),
           correlation.columns)

plt.title("Correlation Heatmap")

plt.savefig("heatmap.png")

plt.close()

# -----------------------------------
# 4. Bar Chart
# -----------------------------------

burnout_count = data["burnout_risk"].value_counts()

plt.figure(figsize=(8, 5))

plt.bar(burnout_count.index.astype(str),
        burnout_count.values,
        label="Burnout Count")

plt.title("Burnout Risk Count")
plt.xlabel("Burnout Risk")
plt.ylabel("Count")

plt.legend()

plt.savefig("bar_chart.png")

plt.close()

# -----------------------------------
# 5. Boxplot
# -----------------------------------

plt.figure(figsize=(8, 5))

plt.boxplot(data["stress_level"])

plt.title("Stress Level Boxplot")
plt.ylabel("Stress Level")

plt.savefig("boxplot.png")

plt.close()

print("All visualizations saved successfully!")
```
